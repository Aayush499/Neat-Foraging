```python
#check if all input nodes influence the output
for i in range(6):
    test_input = [0.0]*6
    test_input[i] = 1.0
    output = net.activate(test_input)
    print(f"Input {test_input} produces output {output}")

baseline_input = [0.0]*6
baseline_output = net.activate(baseline_input)
print(f"Baseline input {baseline_input} produces output {baseline_output}")
```

```python
#assign different values to input 3 and see how output changes
#try 100 values between 0 and 20, print when output changes from "[0.0, 0.0, 0.0, 0.0]"
for i in range(100):
    test_input = [0.0]*6
    test_input[3] = i * 0.2  # Simulate an obstacle detected by the front sensor
    output = net.activate(test_input)
    if output != [0.0, 0.0, 0.0, 0.0]:
        print(f"Input {test_input} produces output {output}")
        
```
